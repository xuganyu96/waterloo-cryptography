# Problem 1
Recall that a distortion map $\phi: E \rightarrow E$ satisfies that $\phi(nP) = n\phi(P)$ for all $P \in E[l]$.

Given $P, aP, bP, cP$, we can pair $P$ with $cP$, $aP$ with $bP$, and compute the modified Weil pairing of each pair:

$$
\begin{aligned}
\hat{e}_l(aP, bP) &= e_l(aP, \phi(bP)) \\
&= e_l(aP, b\cdot \phi(P)) \\
&= e_l(P, \phi(P))^{ab} \\
\hat{e}_l(P, cP) &= e_l(P, \phi(cP)) \\
&= e_l(P, c\phi(P)) \\
&= e_l(P, \phi(P))^c
\end{aligned}
$$

Divide the first pairing by the second:

$$
\frac{\hat{e}_l(aP, bP)}{\hat{e}_l(P, cP)} = e_l(P, \phi(P))^{ab - c}
$$

Recall that $P$ is a point of finite order $l$, so $(ab)P = cP$ if and only if $ab \equiv c \mod l$. On the other hand, because $\phi$ is a distortion map, $e_l(P, \phi(P))^{ab - c} = 1$ if and only if $ab - c \equiv 0 \mod l$, which is equivalent to $ab \equiv c \mod l$. In other words, $e_l(P, \phi(P))^{ab - c} = 1$ if and only if $(ab)P = cP$.

Hence we can solve the ECDDH problem using the distortion map and modified Weil pairing by computing the fraction $\frac{\hat{e}_l(aP, bP)}{\hat{e}_l(P, cP)}$ and checking if it is equal to $1$. This fraction is 1 if and only if $(ab)P = cP$.

# Problem 2

## a)
To prove that $\phi(P) \in E(\mathbb{K})$ it suffices to show that $(\beta x, y)$ satisfies the elliptic curve. 

For $P = \mathcal{O}$, this statement is trivial, because $\phi(\mathcal{O}) = \mathcal{O}$ is trivially on the elliptic curve. 

Now assuming that $P = (x, y) \neq \mathcal{O}$, substitute $\beta x$ for $x$ on the right hand side:

$$
\begin{aligned}
\text{R.H.S.} &= (\beta x)^3 + 1 \\
&= \beta^3 x^3 + 1 \\
&= 1^3 x^3 + 1 \\
&= x^3 + 1 \\
&= y^2 = \text{L.H.S.} \\
\end{aligned}
$$

Thus we've shown that $\phi(P)$ is also on the curve.

## b)
We claim that $\phi(-P) = -\phi(P)$. This statement trivially true for $P = \mathcal{O}$ since $-\mathcal{O} = \mathcal{O}$. On the other hand, if $P = (x, y) \neq \mathcal{O}$, then $\phi(-P) = \phi(x, -y) = (\beta x, -y) = -(\beta x, y) = - \phi(P)$.

We will prove the addition law, first by discussing a few trivial edge cases, then showing the general case.

Trivial case 1: If either of $P_1, P_2$ is $\mathcal{O}$, then the addition law is trivially held: $\phi(P_1 + \mathcal{O}) = \phi(P_1) = \phi(P_1) + \mathcal{O} = \phi(P_1) + \phi(\mathcal{O})$.

Trivial case 2: if $P_1 = -P_2$, then $\phi(P_1 + P_2) = \phi(\mathcal{O}) = \mathcal{O} = \phi(P_1) - \phi(P_1) = \phi(P_1) +\phi(-P_1) = \phi(P_1) + \phi(P_2)$, so the addition law still holds.

Now we can assume that neither $P_1, P_2$ is $\mathcal{0}$ and that $P_1 \neq -P_2$. Let $P_1 = (x_1, y_1)$, $P_2 = (x_2, y_2)$. Recall the elliptic curve addition formula for $P_3 = (x_3, y_3) = P_1 + P_2$:

$$
\begin{cases}
x_3 = \lambda^2 - x_1 - x_2 \\
y_3 = \lambda(x_1 - x_3) - y_1
\end{cases}
$$

Where $\lambda$ is the slope of the line crossing through the two points, or the tagent line if $P_1 = P_2$. For convenience, we further denote the slope of the line/tagent line running through $\phi(P_1), \phi(P_2)$ by $\lambda_\phi$, and the denote the point $\phi(P_1) + \phi(P_2)$ by $P_3^\prime = (x_3^\prime, y_3^\prime)$

If $P_1 \neq P_2$, then $\lambda = (y_2 - y_1)(x_2 - x_1)^{-1}$, and $\lambda_\phi = (y_2 - y_1)(\beta x_2 - \beta x_1)^{-1} = \lambda\beta^{-1} = \lambda\beta^2$. If $P_1 = P_2$, then $\lambda = (3x_1^2 + A)(2y_1)^{-1} = (3x_1^2)(2y_1)^{-1}$, and $\lambda_\phi = (3(\beta x)^2 + A)(2y_1)^{-1} = \beta^2(3x_1^2)(2y_1)^{-1} = \lambda\beta^2$. Therefore, **regardless of whether it is a tagent line or a slope**:

$$
\lambda_\phi = \lambda\beta^2
$$

We can now prove the addition law by proving $\phi(P_3) = P_3^\prime$ for each coordinate:

$$
\begin{aligned}
x(P_3^\prime) &= x(\phi(P_1) + \phi(P_2)) \\
&= \lambda_\phi^2 - x(\phi(P_1)) - x(\phi(P_2)) \\
&= (\lambda\beta^2)^2 - \beta x_1 - \beta x_2 \\
&= \lambda^2\beta^4 - \beta x_1 - \beta x_2 \\
&= \lambda^2\beta - \beta x_1 - \beta x_2 \\
&= \beta(\lambda - x_1 - x_2) \\
&= x(\phi(P_1 + P_2)) \\
&= x(\phi(P_3)) \\
y(P_3^\prime) &= y(\phi(P_1) + \phi(P_2)) \\
&= \lambda_\phi(\beta x_1 - \beta x_3) - y_1 \\
&= \lambda\beta^2 \cdot \beta(x_1 - x_3) - y_1 \\
&= \lambda(x_1 - x_3) - y_1 \\
&= y(P_3) \\
&= y(\phi(P_3))
\end{aligned}
$$

Thus $\phi(P_3)$ and $P_3^\prime$ are coordinate-wise equal.

# Problem 3

## a)
Suppose that $x \in \mathbb{F}_p$ is a primitive cubic roots, then $x^2$ cannot be $1$ because $x^2 \equiv x^3 \equiv 1 \mod p$ implies $x \equiv 1 \mod p$. Thus we know that the (multiplicative) order of $x$ is 3.

By Fermat's little theorem we also know that $x^{p-1} \equiv 1 \mod p$, which means that $p - 1$ must be a multiple of 3, but we know $p \equiv 2 \mod 3$, so $p-1 \equiv 1 \mod 3$ could never be a multiple of 3. Therefore, we've arrived at a contradiction, so primitive cubic root cannot exist in $\mathbb{F}_p$ for $p \equiv 2 \mod 3$.

To prove that $\mathbb{F}_{p^2}$ has primitive cube, we'll use the following two facts:

1. the multiplicative group $\mathbb{F}^\times_{p^2}$ is a finite cyclic group
2. if $d$ divides the order of some cyclic group $G$, then there exists a sub-group of order $d$

Observe that the multiplicative group $\mathbb{F}^\times_{p^2}$ is formed by taking all the non-zero elements, so the order of the multiplicative group is exactly:

$$
\vert\mathbb{F}^\times_{p^2}\vert = p^2 - 1
$$

Because $p \equiv 2 \mod 3$, we know 3 divides $p^2 - 1$, which means that there exists cyclic subgroup of $\mathbb{F}^\times_{p^2}$ with order 3, and the generator of such cyclic subgroup of order 3 would be a primitive cube root.

## b)
From problem 2 part (b) we have already proved the addition law $\phi(P + Q) = \phi(P) + \phi(Q)$. From the addition law we can deduce $\phi(nP) = \phi(P + (n-1)P) = \phi(P) + \phi((n-1)P) = \ldots = n\phi(P)$, thus satisfying the first condition of a distortion map.

To prove that $e_l(P, \phi(P))$ is a primitive $l$-th root, we use proposition 6.50 from the textbook, which states that $e_l(P, \phi(P))$ is a primitive $l$-th root if and only if $P$ is not $\mathcal{O}$ and $\phi(P)$ is not a multiple of $P$. Since $P \in E[l]$ for some $l \geq 5$, it's trivially true that $P \neq \mathcal{O}$. It remains to prove that $\phi(P)$ is not a multiple of $P$.

Suppose that $\phi(P)$ is a multiple of $P$. Denote the coordinates of $P$ by $(x, y)$ for some $x, y \in \mathbb{F}_p$, then the coordinates of any multiples of $P$ must also be in $\mathbb{F}_p$. However, since we proved in part (a) that $\beta \not\in \mathbb{F}_p$, $\phi(P) = (\beta x, y)$'s coordinates are in $\mathbb{F}_p$ if and only if $x = 0$, so $P$ must be $(0, \pm 1)$.

Observe the computation of $P + P$:

- The tangent line's slope is $\lambda = (3x^2)(2y)^{-1} = 0$
- $x' = \lambda^2 - x - x = 0$
- $y' = \lambda(x - x_3) - y = \mp 1$

In other words, $P + P = -P$, so $P$ has order 3, which contradicts the fact that $P \in E[l]$ for prime $l \geq 5$. Therefore, $\phi(P)$ is not a multiple of $P$, and $e_l(P, \phi(P))$ is indeed a primitive $l$-th root.

# Problem 4
First we recall the Boneh-Franklin cryptosystem (and establish some notations):

1. $\mathbb{F}_q, E, l, P, H_1, H_2, B, \phi, \hat{e}_l\leftarrow \text{PGen}(1^\lambda)$ where
    - $q, l$ are prime numbers
    - $E$ is an elliptic curve over the field $\mathbb{F}_q$
    - $P \in E(\mathbb{F}_q)[l]$ is a point of finite order $l$
    - $H_1: \{\text{User IDs}\} \rightarrow E[l]$ is a hash function
    - $H_2: \mathbb{F}_q^\times \rightarrow \{0, 1\}^B$ is another hash function
    - $B$ is the bit-length of the message space
    - $\phi$ is a distortion map and $\hat{e}_l$ is the modified Weil-pairing based on $\phi$
2. $\text{pk}_\text{auth}, \text{sk}_\text{auth} \leftarrow \text{AuthKeyGen}$ generates the authority's key pair, where
    - $s \leftarrow \mathbb{Z}_l$ is a random integer
    - $\text{sk} \leftarrow s$ is the authority's secret key
    - $\text{pk} \leftarrow sP$ is the authority's public key
3. $\text{pk}_\text{Alice}, \text{sk}_\text{Alice} \leftarrow \text{UserKeyGen}$ generates the user's key pair, where
    - Alice's public key is its user ID
    - $P_A \leftarrow H_1(\text{Alice's user ID})$
    - $\text{sk}_\text{Alice} \leftarrow sP_A$ is Alice's secret key. This secret key is computed by the authority and securely delivered to Alice

To encrypt a message $m \in \{0, 1\}^B$, sample a random integer $r \leftarrow \{2, 3, \ldots, q-1\}$ and compute:

$$
\begin{cases}
c_1 = rP \\
c_2 = m \oplus H_2(\hat{e}_l(P_A, sP)^r)
\end{cases}
$$

$c = (c_1, c_2)$ is the encryption.

First note that **regardless of the choice of hash function, the Boneh-Franklin cryptosystem is OW-CCA2 insecure**. This is because given a valid ciphertext $c = (c_1, c_2)$, $c' = (c_1, m' \oplus c_2)$ is also a valid ciphertext. When the adversary receives the challenge ciphertext $c^\ast = (c_1^\ast, c_2^\ast)$, it can generate some random non-zero bit string $m'$ and issue a decryption query on $(c_1^\ast, m' \oplus c_2^\ast)$. The decryption will be $m' \oplus m^\ast$, where $m^\ast$ is the challenge plaintext, and since the adversary knows $m'$, it can recover $m^\ast$.

For the remainder of this discussion, we focus on what happens when $H_1$ returns a multiple of the base point $P$ instead of a truly random point of finite order $l$. However, it is not exactly clear whether $H_1$ returns an opaque multiple $kP$ without revealing or a clear multiple $kP$ that reveals the $k$, so we will discuss the two cases separately:

1. $H_1$ returns an opaque multiple $kP$ without giving away $k$
2. $H_1$ directly returns $k$

Note that these two cases are distinct because recovering $k$ from $P, kP$ is solving the EC discrete log problem, which is conjectured to be hard.

## 1)
If $H_1$ does not give away the discrete log, then I suspect that Boneh-Franklin's IBE is IND-CPA secure. This is because $c_1$ is independent of the message, while in $c_2$, $H_2$ functions as a pseudorandom one-time pad. Assuming $H_2$ to be collision resistant, the only way to distinguish outputs of $H_2$ is to be able to compute the modified Weil pairing:

$$
\hat{e}_l(P_A, sP)^r = \hat{e}_l(P, P)^{ksr}
$$

from $kP, sP, rP$, but this is precisely the bilinear Diffie-Hellman problem, which is conjectured to be computationally infeasible.

## 2)
Now suppose that instead of outputing a random point of finite order $P_a \leftarrow E[l]$, $H_1$ outputs a random integer $k \leftarrow \{1, 2, \ldots, l-1\}$, and the authority/encryptor computes Alice's public base point $P_A \leftarrow kP$. In this scenario, this cryptosystem is **one-way ciphertext-only-attack insecure**, meaning that an adversary can decrypt arbitrary message using only the public keys and the ciphertext.

Recall from previous argument, if the adversary can compute $\hat{e}_l(P, P)^{ksr}$, then it can apply $H_2$ and obtain the mask. The adversary can then use the mask to recover the message:

$$
m = c_2 \oplus H_2(\hat{e}_l(P_A, sP)^r) = c_2 \oplus H_2(\hat{e}_l(kP, sP)^r) = c_2 \oplus H_2(\hat{e}_l(P, P)^{ksr})
$$

Indeed if $H_1$ outputs $k$, then the adversary can use $rP$ (which is the first component of the ciphertext and thus public), $sP$ (authority's public key) and $k$ (adversary can obtain $k$ by feeding Alice's user ID into $H_1$) to compute $\hat{e}_l(P, P)^{ksr}$:

$$
\hat{e}_l(rP, sP)^k = \hat{e}_l(P, P)^{ksr}
$$

Therefore, the adversary can decrypt arbitrary ciphertext.
