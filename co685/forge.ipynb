```python
from lib685.dsa import Params, KeyPair, Signature, sign_hashed, verify_hashed

p, q, g = 48731, 443, 5260
sk_val = 242
m_hashed = 343
nonce = 427

params = Params(p, q, g)
keypair = KeyPair(pow(g, sk_val, p), sk_val, params)
sigma = sign_hashed(keypair, m_hashed, nonce)
print(verify_hashed(keypair.pk, m_hashed, sigma, True))
```
