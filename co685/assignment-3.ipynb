## a)
For each of the prime $p \in \{3, 11, 17\}$, if the integer $a$ is not co-prime with $p$, then $p \mid a$, and the congruences are trivially true because both sides equal to 0. For the remainder of this argument, for each of the possible prime $p$, we assume $a$ to be co-prime with $p$, which means that Fermat's little theorem holds.

For $p = 3$, $a^2 \equiv 1 \mod 3$:

$$
\begin{aligned}
a^{561} &\equiv a^{2 \cdot 280 + 1} \mod 3 \\
&\equiv (a^2)^{280} \cdot a \mod 3 \\
&\equiv 1^{280} \cdot a \mod 3 \\
&\equiv a \mod 3
\end{aligned}
$$

For $p = 11$, we have $a^{10} \equiv 1 \mod 11$

$$
\begin{aligned}
a^{561} &\equiv a^{10 \cdot 56 + 1} \mod 11 \\
&\equiv (a^{10})^{56} \cdot a \mod 11 \\
&\equiv 1^{56} \cdot a \mod 11 \\
&\equiv a \mod 11
\end{aligned}
$$

For $p = 17$, we have $a^{16} \equiv 1 \mod 17$:

$$
\begin{aligned}
a^{561} &\equiv a^{16 \cdot 35 + 1} \mod 17 \\
&\equiv (a^{16})^{35} \cdot a \mod 17 \\
&\equiv 1^{35} \cdot a \mod 17 \\
&\equiv a \mod 17
\end{aligned}
$$

Thus we've shown that the three individual congruences hold for all integers $a$.

The fact that the three prime congruences imply the composite congruences is a result of the Chinese remainder theorem. First we take $p = 3$ and $p = 11$ and write out the congruence as Euclidean division:

$$
a^{561} = m \cdot 3 + a = n \cdot 11 + a
$$

Where $m, n$ are integers. Removing the remainder $a$ we have $m \cdot 3 = n \cdot 11$, which implies that $11 \mid m$. Taking the congruences wiht $p = 3$ and $p = 17$ we can also deduce that $17 \mid m$. Therefore, $m = 11 \cdot 17 \cdot k$ for some other integer $k$, and we can rewrite the Euclidean division above as:

$$
a^{561} = k \cdot 11 \cdot 17 \cdot 3 + a = k \cdot 561 + a
$$

Thus we have $a^{561} \equiv a \mod 561$

## b)
We will prove that the product of two distinct primes cannot be a Carmichael number by showing that **if a Carmichael number is the product of two primes, then the two primes must be equal**.

Let $n = p \cdot q$ be a Carmichael number that is the product of two primes $p, q$. By the definition of a Carmichael number we know that $a^{p \cdot q} \equiv a \mod n$ for all integers a.

By Chinese remainder theorem, we know that $a^{pq} \equiv a \mod p$ for all integers $a$. This means that this congruence hold for the primitive root(s) of $p$, as well. From assignment 1 we know that because $p$ is prime, at least one primitive root exists, and we will denote it by $g$. By the definition of primitive roots we know that the order of $g$ is $p-1$.

By Fermat's little theorem we know that $g^p \equiv g \mod p$, so $g^{pq} \equiv g^q \equiv g \mod p$. Because $g$ is a primitive root and thus has a multiplicative inverse, we can multiply $g^{-1}$ on the congruence to get $g^{q-1} \equiv 1 \mod p$. From assignment 1 we know that the order of $g$ must thus divide $q-1$, which means that $p - 1 \mid q - 1$.

Swapping $p, q$ in the two paragraphs above, we can also deduce that $q - 1 \mid p - 1$. This necessarily means that $p - 1 = q - 1$, so $p = q$.

We begin by deriving a number of lemmas before proving the actual results.

**(Lemma 1)** if $a, b$ are odd integers, then 

$$
\frac{a-1}{2} + \frac{b-1}{2} \equiv \frac{ab-1}{2} \mod 2
$$

**Proof**: First observe that $a-1, b-1$ are both even, so $(a-1)(b-1) \equiv 0 \mod 4$. Rearranging this congruence we have $ab -1 \equiv (a - 1) + (b - 1) \mod 4$, which implies that $ab - 1 = 4k + (a - 1) + (b-1)$ for some integer $k$. Divide both sides by 2 we have

$$
\frac{ab-1}{2} = 2k + \frac{a-1}{2} + \frac{b-1}{2}
$$

This proves Lemma 1.

**(Lemma 2)** if $p_1, p_2, \ldots, p_t$ are odd integers, then

$$
\sum_{i=1}^t \frac{p_i - 1}{2} \equiv \frac{(\prod_{i=1}^t p_i) - 1}{2} \mod 2
$$

**Proof**: we will prove by induction. The base case where $t = 1$ is trivial. Now suppose the congruence to hold for $t = n$, then

$$
\begin{aligned}
\sum_{i=1}^{n+1} \frac{p_i - 1}{2} &= \sum_{i=1}^{n} \frac{p_i - 1}{2} + \frac{p_{n+1}-1}{2} \\
&\equiv \frac{(\prod_{i=1}^n p_i) - 1}{2} + \frac{p_{n+1}-1}{2} \mod 2
\end{aligned}
$$

Because the product of odd integers is an odd integer, we can apply Lemma 1 to RHS:

$$
\frac{(\prod_{i=1}^n p_i) - 1}{2} + \frac{p_{n+1}-1}{2} = \frac{(\prod_{i=1}^{n+1} p_i) - 1}{2}
$$

Thus we have proved the inductive case.

**(Lemma 3)** if $a, b$ are odd integers, then

$$
\frac{a^2b^2 - 1}{8} \equiv \frac{a^2 - 1}{8} + \frac{b^2 - 1}{8} \mod 2
$$

**Proof:** First observe that for odd integer $a$, $a^2 - 1$ is divisible by 4. This is because $a^2 - 1 = (a+1)(a-1)$, where $a+1, a-1$ are both even, and the product of two even numbers is guaranteed to be divisible by 4. Therefore we know that $a^2 - 1 \equiv b^2 - 1 \equiv 0 \mod 4$, which means that $a^2 - 1 = 4k, b^2 - 1 = 4l$ for some integers $k, l$. Multiply the two equations, we get $(a^2 - 1)(b^2 - 1) = 16kl$, therefore:

$$
(a^2 - 1)(b^2 - 1) \equiv 0 \mod 16
$$

Rearranging the congruence above:

$$
a^2b^2 - 1 \equiv (a^2 - 1) + (b^2 - 1) \mod 16
$$

Which implies that for some integer $k$:

$$
a^2b^2 - 1 = 16k + (a^2 - 1) + (b^2 - 1)
$$

Divide both sides by 8, we get:

$$
\frac{a^2b^2 - 1}{8} = 2k + \frac{a^2 - 1}{8} + \frac{b^2-1}{8}
$$

This proves the lemma

**Lemma 4**: let $p_1, p_2, \ldots, p_t$ be odd integers, then

$$
\sum_{i=1}^{t}\frac{p_i^2 - 1}{8} \equiv \frac{(\prod_{i=1}^{t}p_i^2) - 1}{8} \mod 2
$$

**Proof:** again we will prove by induction, and again the base case $t = 1$ is trivial. Now suppose the congruence hold for $t = n$, then 

$$
\begin{aligned}
\sum_{i=1}^{n+1}\frac{p_i^2 - 1}{8} &\equiv \sum_{i=1}^{n}\frac{p_i^2 - 1}{8} + \frac{p_{n+1}^2 - 1}{8} \mod 2 \\
&\equiv \frac{(\prod_{i=1}^{n}p_i^2) - 1}{8} + \frac{p_{n+1}^2 - 1}{8} \mod 2 \\
&\equiv \frac{(\prod_{i=1}^{n}p_i)^2 - 1}{8} + \frac{p_{n+1}^2 - 1}{8} \mod 2
\end{aligned}
$$

Apply Lemma 3 to RHS:

$$
\frac{(\prod_{i=1}^{n}p_i)^2 - 1}{8} + \frac{p_{n+1}^2 - 1}{8} \equiv \frac{(\prod_{i=1}^{n+1}p_i^2) - 1}{8} \mod 2
$$

Thus we've proved Lemma 4.

Last but not least we claim a trivial result: $(-1)^a = (-1)^b$ if and only if $a \equiv b \mod 2$. For the remainder of the writeup, these four lemmas and the trivial result will be directly used without explicit declaration.

## c)
First we claim a trivial result:

Let $p$ be a prime number, then the Legendre symbol of a product is the product of Legendre symbol $(\frac{ab}{p}) = (\frac{a}{p})(\frac{b}{p})$. For $a, b$ that are co-prime with $p$, this can be proved using Euler's criterion: $(\frac{a}{p})(\frac{b}{p}) \equiv a^\frac{p-1}{2}b^\frac{p-1}{2} \equiv (ab)^\frac{p-1}{2} = (\frac{ab}{p})$; if any of $a, b$ is not co-prime with $p$, then both sides are trivially 0

Now suppose $m, n$ has prime factorization:

$$
m = p_1p_2\ldots p_t \\
n = q_1q_2 \ldots q_s
$$

Then

$$
\begin{aligned}
\bigg(\frac{m}{n}\bigg) &= \bigg(\frac{\prod_{i=1}^tp_i}{\prod_{j=1}^s q_j}\bigg) \\
&= \prod_{i=1}^t\prod_{j=1}^s\bigg(\frac{p_i}{q_j}\bigg)
\end{aligned}
$$

Similarly we have

$$
\begin{aligned}
\bigg(\frac{n}{m}\bigg) &= \bigg(\frac{\prod_{j=1}^s q_j}{\prod_{i=1}^tp_i}\bigg) \\
&= \prod_{i=1}^t\prod_{j=1}^s\bigg(\frac{q_j}{p_i}\bigg)
\end{aligned}
$$

Multiply the two equations:

$$
\begin{aligned}
\bigg(\frac{m}{n}\bigg)\bigg(\frac{n}{m}\bigg) &= \prod_{i=1}^t\prod_{j=1}^s\bigg(\frac{p_i}{q_j}\bigg) \cdot \prod_{i=1}^t\prod_{j=1}^s\bigg(\frac{q_j}{p_i}\bigg) \\
&= \prod_{i=1}^t\prod_{j=1}^s\bigg(\frac{p_i}{q_j}\bigg)\bigg(\frac{q_j}{p_i}
\bigg) \\
&= \prod_{i=1}^t\prod_{j=1}^s(-1)^\frac{(p_i - 1)(q_j - 1)}{4} \\
&= (-1)^{\sum_{i=1}^t\sum_{j=1}^s\frac{(p_i - 1)(q_j - 1)}{4}}
\end{aligned}
$$

Observe the exponent:

$$
\begin{aligned}
\sum_{i=1}^t\sum_{j=1}^s\frac{(p_i - 1)(q_j - 1)}{4} &= \sum_{i=1}^t (\frac{p_i - 1}{2} \sum_{j=1}^s\frac{q_j - 1}{2}) \\ 
&\equiv \sum_{i=1}^t(\frac{p_i-1}{2} \cdot \frac{q_1q_2 \ldots q_s - 1}{2}) \mod 2 \\
&\equiv \sum_{i=1}^t(\frac{p_i-1}{2} \cdot \frac{n - 1}{2}) \mod 2 \\
&\equiv \frac{n - 1}{2} \sum_{i=1}^t(\frac{p_i-1}{2}) \mod 2 \\
&\equiv \frac{n - 1}{2} \frac{p_1p_2 \ldots p_t - 1}{2} \mod 2 \\
&\equiv \frac{n - 1}{2} \frac{m - 1}{2} \mod 2 \\
&\equiv \frac{(m-1)(n-1)}{4} \mod 2
\end{aligned}
$$

Plugging this into the RHS of the previous equation:

$$
\bigg(\frac{m}{n}\bigg)\bigg(\frac{n}{m}\bigg) = (-1)^{\frac{(m-1)(n-1)}{4}}
$$
