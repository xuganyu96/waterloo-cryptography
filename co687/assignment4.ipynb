$$
r_1e_1 + r_2e_2 = 1
$$

Recall that the ciphertexts of the common message $m$ under the two exponents are $c_1 \equiv m_1^{e_1} \mod n$ and $c_2 \equiv m_2^{e_2} \mod n$. The adversary can raise the two ciphertexts to $r_1$ and $r_2$ exponents respectively, then multiply them together:

$$
\begin{aligned}
c_1^{r_1} \cdot c_2^{r_2} &\equiv (m^{e_1})^{r_1} \cdot (m^{e_2})^{r_2} \mod n \\
&\equiv m^{r_1e_1 + r_2e_2} \mod n \\
&\equiv m \mod n
\end{aligned}
$$

Thus the adversary has obtained the message.

3. $\mathcal{A}_\text{DDH}$ checks whether $h$ is equal to $z$. If they are equal, $\mathcal{A}_\text{DDH}$ claims $z$ to be $g^{ab}$; otherwise $\mathcal{A}_\text{DDH}$ claims $z$ to be a truly random element from $G$.

First observe that for Diffie-Hellman parameters $q, G, g$ where $G$ is a cyclic group of order $q$ and $g$ is a generator of the group, arbitrary element $x \in G$ is a perfect square if and only if $x = g^e$ for some even integer $e$.

In the forward direction, this is true because if $x$ is a perfect square, then there exists a second element $y = g^d$ such that $x = y^2 = g^{2d}$, so $e = 2d$ is even. In the backward direction, if $e$ is even then there exists integer $d$ such that $e = 2d$, so $x = g^e = (g^d)^2$, meaning that $g^d$ is a square root of $x$.

Now observe that if $a, b$ are random integers, then $ab$ has a 75% chance of being even, because $ab$ is odd if and only if $a, b$ each is odd. On the other hand, a truly random integer $c$ has exactly 50% of being even. This means that if $g$ is not a perfect square, then $g^{ab}$ has a 75% chance of being a perfect square, while $g^c$ has a 50% chance of being a perfect square.

The advantage of such an adversary is $0.5 \cdot 0.75 + 0.5 \cdot 0.5 - 0.5 = 0.125$, which is definitely not negligible.

## a)
When discussing IND-CPA and IND-CCA games in a public-key cryptosystem, an encryption oracle is not necessary because the adversary is assumed to have access to the public key and can thus perform the encryption procedure by itself.

## b)
The IND-CPA game of ElGamal is as follows:

1. The challenger generates the parameters $(p, G, g) \leftarrow \text{PGen}(1^\lambda)$ where $G$ is a cyclic group of prime order $p$ and $g$ is a generator of the group $G$.
2. The challenger generates the keypair $(g^x, x) \leftarrow \text{KeyGen}(1^\lambda, (p, G, g))$ where $g^x$ is the public key and $x \in \mathbb{Z}_p$ is the private key
3. The challenger gives all parameters and the public key to the adversary, and the adversary generates a pair of challenge messages $m_0, m_1 \leftarrow \mathcal{A}(1^\lambda, (p, G, g), g^x)$
4. The challenger uniformly sample a bit $b \leftarrow \{0, 1\}$, then encrypt $m_b$ under the generated public key: $c^* \leftarrow E(\text{pk}, m_b)$.
5. The adversary receives the challenge ciphertext and outputs a bit $b^* \leftarrow \mathcal{A}(1^\lambda, (p, G, g), g^x, (m_0, m_1, c^*))$. The adversary wins the game if $b^* = b$

Notice that there is no encryption oracle in the IND-CPA game because the adversary is always given the public key, which it can use to encrypt arbitrary message that it chooses.

The IND-CCA (in fact, IND-CCA2) game of ElGamal is as follows:

1. The challenger generates the parameters $(p, G, g) \leftarrow \text{PGen}(1^\lambda)$ where $G$ is a cyclic group of prime order $p$ and $g$ is a generator of the group $G$.
2. The challenger generates the keypair $(g^x, x) \leftarrow \text{KeyGen}(1^\lambda, (p, G, g))$ where $g^x$ is the public key and $x \in \mathbb{Z}_p$ is the private key
3. The adversary is given the parameters and the public key, as well as access to a decryption oracle. The adversary can adaptively ask the decryption oracle to decrypt arbitrary ciphertext: for $i \in \{1, 2, \ldots, q\}$, $c_i \leftarrow \mathcal{A}(1^\lambda, (p, G, g), g^x, \{(c_j, m_j)\}_{j<i})$, $m_i \leftarrow \text{Decryption Oracle}(c_i)$
4. The adversary generates a pair of challenge messages $m^*_0, m^*_1 \leftarrow \mathcal{A}(1^\lambda, (p, G, g), g^x, \{(m_i, c_i)\}_{i=1}^q)$
4. The challenger uniformly sample a bit $b \leftarrow \{0, 1\}$, then encrypt $m^*_b$ under the generated public key: $c^* \leftarrow E(\text{pk}, m_b)$.
6. The adversary continues to have access to the decryption oracle, although the decryption oracle will refuse to decrypt the challenge ciphertext:  
for $i \in \{q+1, \ldots, q+p\}$, $c_i \leftarrow \mathcal{A}(1^\lambda, (p, G, g), g^x, \{(c_j, m_j)\}_{j<i}, (m^*_0, m^*_1, c^*))$,  
$m_i \leftarrow \text{Decryption Oracle}(c_i)$,  
$\bot \leftarrow \text{Decryption Oracle}(c^*)$
7. The adversary outputs $b^* \leftarrow \mathcal{A}(1^\lambda, (p, G, g), g^x, \{(c_i, m_i)\}_{i=i}^{p+q}, (m^*_0, m^*_1, c^*))$ based on all past queries. The adversary wins if $b^* = b$

The IND-CCA1 game is similar to IND-CCA2 game except adversary cannot access the decryption oracle after receiving the challenge ciphertext.

## c)
Let $\mathcal{A}_\text{DDH}$ denote the DDH adversary and $\mathcal{A}_\text{ElGamal}$ denote the IND-CPA adversary for ElGamal.

In the DDH game, $\mathcal{A}_\text{DDH}$ receives $g, g^a, g^b, h$, where $h$ has a 50% chance of being $g^{ab}$ and 50% chance of being a random element of the group $g^c \in G$. Since $a, b$ are both uniformly randomly sampled, we can use $g^a$ as a public key and $g^b$ as one of the random term in the challenge ciphertext in the IND-CPA game. We then use $h$ as the other random term such that, when $h = g^{ab}$, the challenge ciphertext is well-formed, and the IND-CPA adversary will retain its advantage, and when $h = g^c$, the challenge ciphertext distribution is not affected by the choice of the plaintext, so the IND-CPA adversary will have no advantage. From here, the DDH adversary simply checks whether the IND-CPA adversary's guess is correct or not: if correct, then DDH claims $h = g^{ab}$, else the DDH adversary claims $h = g^c$.

Here is the description of $\mathcal{A}_\text{DDH}$'s actions, assuming that it has access to $\mathcal{A}_\text{IND-CPA}$, an IND-CPA adversary.

1. When $\mathcal{A}_\text{DDH}$ receives $p, G, g, g^a, g^b, h \leftarrow \{g^{ab}, g^c\}$ from the DDH challenger, it gives $p, G, g$ and $g^a$ to $\mathcal{A}_\text{IND-CPA}$ and receives the chosen messages $m_0, m_1 \leftarrow \mathcal{A}_\text{IND-CPA}(1^\lambda, (p, G, g), g^a)$.
2. $\mathcal{A}_\text{DDH}$ samples a random bit $i \leftarrow \{0,1\}$ and gives $c^* = (g^b, m_b \cdot h)$ to $\mathcal{A}_\text{IND-CPA}$
3. The IND-CPA adversary outputs a guess $i^* \leftarrow \mathcal{A}_\text{IND-CPA}(1^\lambda, (p, G, g), g^a, (m_0, m_1, c^*))$
4. $\mathcal{A}_\text{DDH}$ claims $h = g^{ab}$ if $i^* = i$, else it claims $h = g^c$

There are a few things to notes:

- The challenge ciphertext $c^* = (g^b, m_b \cdot h)$ is statististically identically distributed as a truly random ElGamal ciphertext $c = (g^r, m \cdot g^r)$, because the both terms are truly random element of $G$.
- When $h = g^{ab}$, the challenge ciphertext is a valid encryption of one of the challenge messages, so the IND-CPA adversary will retain its advantage
- When $h = g^c$, the challenge ciphertext's distribution is unaffected by the chosen message. $m \cdot g^c$ is a truly random element of $G$ regardless of choice of $m$ since $g^c$ is a truly random element. Therefore, the IND-CPA adversary will have no advantage.

Since there is a 50% chance that $h = g^c$, there is a 50% chance the IND-CPA adversary will retain its advantage. Therefore, the advantage of the DDH adversary is half of the IND-CPA adversary, meaning that if the IND-CPA adversary has non-negligible advantage, then the DDH adversary has non-negligible advantage.


## d)
The ElGamal encryption scheme is not CCA-secure because it exhibits some homomorphic properties. Suppose we have $c = (c_1, c_2) = (g^r, m \cdot (g^x)^r)$ as an encryption of some message $m \in \mathbb{Z}_p$, then $c^\prime = (c_1, 2 \cdot c_2) = (g^r, (2 \cdot m) \cdot (g^x)^r)$ is a valid encryption of the message $2 \cdot m$.

This means that we can build an IND-CCA2 adversary: after receiving the challenge ciphertext $c^* = (c^*_1, c^*_2)$, the adversary can query the "double ciphertext" $(c^*_1, 2 \cdot c^*_2)$ and obtain $2 \cdot m_b$ as the output because the "double ciphertext" is distinct from the challenge ciphertext. From there it is trivial to recover the challenge message $m_b$, and the IND-CCA2 adversary will be able to distinguish the ciphertext.

# Problem 5
Here is my implementation of an elliptic curve. The group operation of points on an elliptic curve is described on Wikipedia: https://en.wikipedia.org/wiki/Elliptic_curve_point_multiplication.

```python
from __future__ import annotations

class EllipticCurve:
    """(finite field) elliptic curve y^2 = x^3 + ax + b
    """
    def __init__(self, p, a, b):
        self.p = p
        self.a = a
        self.b = b
    
    def validate(self, x: int, y: int):
        """Return True iff and point is on the curve"""
        lhs = (y * y) % self.p
        rhs = ((x ** 3) + self.a * x + self.b) % self.p
        return lhs == rhs

    def solve_y(self, x: int):
        """Given a value of x, solve for y with brute-force. Always returns
        a pair of solution if it exists, otherwise return None
        """
        x = x % self.p
        for y in range(self.p):
            remainder = (y ** 2 - x ** 3 - self.a * x - self.b) % self.p
            if remainder == 0:
                return (y, self.p - y)
        return None
    
class Point:
    """A single point on the elliptic curve identified by the coordinate. The
    identity is identified by having no identity
    """
    def _linear_slope(self, other: Point):
        """assume that self and other are two points such that non of them is
        identity, they are not identical, and they are not negations of each
        other, then the slop of the line can be linearly computed
        """
        slope = pow(other.x - self.x, -1, self.curve.p)
        slope = (other.y - self.y) * slope
        slope = slope % self.curve.p
        return slope

    def _tagent_slope(self):
        """The slope of the tagent line cutting through this point"""
        slope = pow(2 * self.y, -1, self.curve.p)
        slope = (3 * self.x * self.x + self.curve.a) * slope
        return slope % self.curve.p

    @staticmethod
    def identity(curve: EllipticCurve):
        return Point(None, None, curve)
    
    def is_identity(self):
        return self.x is None
    
    def negation(self):
        if self.is_identity():
            return Point.identity(self.curve)
        return Point(self.x, -self.y + self.curve.p, self.curve)

    def __init__(self, x, y, curve: EllipticCurve):
        if x is not None and not curve.validate(x, y):
            raise ValueError("Point is not on the curve")
        self.x = x
        self.y = y
        self.curve = curve
    
    def __eq__(self, other):
        """Two points are equal if they are both identity or if they are
        coordinate-wise equal
        """
        if not isinstance(other, Point):
            raise TypeError("Equality with non-point not defined")
        return (
            (self.is_identity() and other.is_identity()) 
            or (self.x == other.x and self.y == other.y)
        )
    
    def __add__(self, other):
        if not isinstance(other, Point):
            raise TypeError("Addition with non-point not defined")
        
        # If one of the operand is identity then return the other
        if self.is_identity():
            return other
        if other.is_identity():
            return self
        
        # P + (-P) = 0
        if self == other.negation():
            return Point.identity(self.curve)
        
        # With P + Q where P, Q are not identity and not x-axis mirrors, there
        # is a common formula, although there is a slope term that is evaluated
        # differently depending on P == Q or P != Q
        slope = (
            self._tagent_slope() 
            if self == other else self._linear_slope(other)
        )
        x = ((slope * slope) - other.x - self.x) % self.curve.p
        y = (-slope * x - self.y + slope * self.x) % self.curve.p
        return Point(x, y, self.curve)
    
    def __mul__(self, other):
        if not isinstance(other, int):
            raise TypeError("Non-scaler multiplication not defined")
        prod = Point(self.x, self.y, self.curve)
        for _ in range(other-1):
            prod += Point(self.x, self.y, self.curve)
        return prod
        

    def __repr__(self):
        if self.is_identity():
            return "<Point 0>"
        return f"<Point ({self.x}, {self.y})>"
```

```python
# a)
class ECPRG:
    """The toy pseudorandom generator"""
    def __init__(self, curve: EllipticCurve, p1: Point, p2: Point, start: int):
        self.curve = curve
        self.p1 = p1
        self.p2 = p2
        self.start = start
    
    def generate(self, limit: int | None = None):
        s_prev = self.start
        count = 1
        while limit is None or (count <= limit):
            r = (self.p1 * s_prev).x
            s_prev = (self.p1 * r).x
            t = (self.p2 * r).x
            yield t
            count += 1

curve = EllipticCurve(19, 2, 3)
p1 = Point(1, 14, curve)
p2 = Point(3, 13, curve)

prg = ECPRG(curve, p1, p2, 2)
for t in prg.generate(3):
    print(t)
```

## b)
First we give the algorithm, then we will explain why it works.

```
1. Solve the elliptic curve equation with x = t_1. Among the two solutions, pick one of them.
2. Compute c * X, where X is the chosen solution from part 1
3. s_1 is the x-coordinate of c * X
```

Given $t_1 = x(r_1Q)$, we know the x-coodinate of the point $r_1Q$. From here we can try to solve the elliptic curve. By the fundamental theorem of algebra we know there can be up to two distinct solutions, and knowing that the only $y$ term is a qudratic term, we know that if there are two solutions, then they are mirrors of each other across x-axis. Denote the two solutions by $X_1, X_2$, then we know one of them is exactly $r_1Q$ and $X_1 = -X_2$.

We can substitute $P = cQ$ into the equation for $s_1$:

$$
\begin{aligned}
s_1 &= x(r_1P) \\
&= x(r_1 \cdot (cQ))
\end{aligned}
$$

Because scalar multiplication is defined as repeated group operation within $E(\mathbb{F}_p)$, we know $r_1 \cdot (cQ) = c \cdot (r_1Q)$, so we have

$$
s_1 = x(c \cdot (r_1Q))
$$

We don't know the exact value of $r_1Q$, but we know two candidate values that are inverses of each other. Furthermore, we know that inverses are preserved across scaler multiplication: $-c \cdot X = c \cdot (-X)$. Therefore, it is easy to deduce that regardless of which of $X_1, X_2$ we plug into the equation above, the output value will be the same and be the correct value: $s_1 = x(c \cdot X_1) = x(c \cdot X_2)$. Thus we have predicted the value of $s_1$.

I propose to pick the points $P, Q$ such that linear relationship $P = cQ$ is impossible.

It is possible to choose a curve such that the group of points on the curve is not cyclic. This means that it's possible to have proper non-trivial subgroup. Denote $G \subsetneq E(\mathbb{F}_p)$ to be the non-trivial proper subgroup, then we can pick a point $Q$ from within $G$ and a point $P$ from outside $G$, and we will have $P, Q$ that are guaranteed to not have linear relationship $P = cQ$ because $cQ$ is guaranteed to be in $G$ while $P$ is not.
